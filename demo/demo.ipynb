To use `ChartGenerator` class we need:
<ol>
<li>Parameter Mapping - [json] - specifies tag meanings, eg. "ftfa" means that the chart has a "fantasy" font</li> 
<li>Charts descriptions - [json/dict/list] - charts to generate divided by chart type; can be batched </li> 
</ol>

# Initializing ChartGenerator
To initialize, pass Parameter Mapping document.

````python
g = ChartGenerator("sample_map.json")
````

## Graphing Single Charts
To graph one particular, specify chart type and parameter tuple/list

````python
g.graph("pie", ("CM", "dd12", "sano", "sm0", "ftns", "fs12", "llcl", "pano", "we05", "ro090", "ad11"))
````

````python
g.graph("donut", ("MB", "dd06", "sa00", "sm1", "ftfa", "fs14", "llcr", "pano", "we05", "ro090", "ad11","dh5"))
````

````python
g.graph("bar", ("CL", "dd12", "sano", "sm0", "ftns", "fs12", "llul", "pano", "e10a1", "bw6", "bs000", "boh", "gr0y"))
````

## Generating Unbatched Charts
To generate several charts at once, specify chart type and list of parameter tuples/lists. All images will be saved in the working directory.

````python
%%capture
for k,v in sample_batches["0"].items():
    g.multi_graph(k, v, save_=True)
````

## Generating Batched Charts
To generate several batches of charts at once, specify batching dictionary of the format
```python
sample_batches = {
    "0": {
        "pie": [...],
        "donut": [...],
        "bar": [...]
    },
    ...
}
```
The images will be saved in folders corresponding to the batch, in the working directory.

````python
%%capture
g.generate(sample_batches)
````
